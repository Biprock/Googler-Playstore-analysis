## Data preprocessing

```python
#droppig duplicate rows
df.drop_duplicates(inplace=True)
df
```

### Data cleaning

```python
df["Category"].unique()
```

```python
df[df["Category"]=='1.9']
```

```python
df.loc[10472] = df.loc[10472].shift()
df['App'].loc[10472] = df['Category'].loc[10472]

df['Category'].loc[10472] = np.nan
df.loc[10472]
```

```python
df["Rating"].unique()
```

```python
# Change Datatype Object into float
df["Rating"] = df["Rating"].astype("float")
df["Rating"].dtype
```

```python
df["Reviews"].unique()
```

```python
df["Reviews"] = df["Reviews"].str.replace('3.0M',"3000000")
```

```python

#Change Datatype Object into Float
df["Reviews"] = df["Reviews"].astype("float")
df["Reviews"].dtype
```

```python
df["Size"].unique()
```

```python
df["Size"] = df["Size"].replace('Varies with device',np.NaN)
df["Size"] = df["Size"].str.replace("M","")
df["Size"] = df["Size"].str.replace("k","")
df["Size"] = df["Size"].str.replace("+","")
```

```python
# Change Datatype Object into Float
df["Size"] = df["Size"].astype("float")
df["Size"].dtype
```

```python
df["Installs"].unique()
```

```python
df["Installs"] = df["Installs"].str.replace("+","")
df["Installs"] = df["Installs"].str.replace(",","")
```

```python

# Change Datatype Object into Float
df["Installs"] = df["Installs"].astype("float")
df["Installs"].dtype
```

```python

df["Price"].unique()
```

```python

df["Price"] = df["Price"].str.replace("$","")
```

```python
df["Price"] = df["Price"].astype("float")
df["Price"].dtype
```

```python
df["Last Updated"].unique()
```

```python

df["Last Updated"] = pd.to_datetime(df["Last Updated"])
df["Last Updated"]
```

## EDA(Exploratory Data Analysis)

```python
#word cloud
plt.figure(figsize=(12,7))
wordcloud = WordCloud(background_color="white", width=512,height=384).generate(" ".join(df))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
#heatmap
plt.figure(figsize=(12,7))
sns.heatmap(data=df.corr(),annot=True)
plt.show()
```

```python
#rating distribution
plt.figure(figsize=(15,7))
ax = sns.countplot(data=df,x=df["Rating"])
plt.title(" Distribution Of Rating ", weight = "bold",fontsize = 18,pad=10)
plt.xticks(weight="bold",rotation=90)
plt.yticks(weight="bold")
plt.xlabel("Rating",weight="bold",labelpad = 12,fontsize = 12)
plt.ylabel("Count",weight="bold",labelpad = 12,fontsize = 12)

for i in ax.containers:
    i.datavalues
    ax.bar_label(i,fontweight="bold")
    
plt.show()
```

```python
#category distribution
plt.figure(figsize=(15,7))
ax =sns.countplot(data=df,x=df["Category"])
plt.title(" Distribution Of Category ", weight = "bold",fontsize = 20,pad=10)
plt.xticks(weight="bold",rotation=90)
plt.yticks(weight="bold")
plt.xlabel("Category",weight="bold",labelpad = 12,fontsize = 15)
plt.ylabel("Count",weight="bold",labelpad = 12,fontsize = 12)

for i in ax.containers:
    i.datavalues
    ax.bar_label(i,fontweight="bold")
plt.show()
```

```python
#content rating distribution
plt.figure(figsize=(15,7))
ax =sns.countplot(data=df,x=df["Content Rating"])
plt.title(" Distribution Of Content Rating ", weight = "bold",fontsize = 20,pad=10)
plt.xticks(weight="bold",rotation=90)
plt.yticks(weight="bold")
plt.xlabel("Content Rating",weight="bold",labelpad = 12,fontsize = 15)
plt.ylabel("Count",weight="bold",labelpad = 12,fontsize = 12)

for i in ax.containers:
    i.datavalues
    ax.bar_label(i,fontweight="bold")
plt.show()
```

```python
#installion distribution
plt.figure(figsize=(15,7))
sns.barplot(data=df,x=df["Category"], y = df["Installs"])
plt.title(" Distribution Of Category by Installs ", weight = "bold",fontsize = 20,pad=10)
plt.xticks(weight="bold",rotation=90)
plt.yticks(weight="bold")
plt.xlabel("Category",weight="bold",labelpad = 12,fontsize = 15)
plt.ylabel("Installs",weight="bold",labelpad = 12,fontsize = 12)
plt.show()
```

```python
#reviews distribution
plt.figure(figsize=(15,7))
sns.lineplot(data=df,x=df["Category"], y = df["Reviews"],marker="*")
plt.title(" Distribution Of Category by Reviews ", weight = "bold",fontsize = 20,pad=10)
plt.xticks(weight="bold",rotation=90)
plt.yticks(weight="bold")
plt.xlabel("Category",weight="bold",labelpad = 12,fontsize = 15)
plt.ylabel("Reviews",weight="bold",labelpad = 12,fontsize = 12)
plt.show()
```
